```python
import matplotlib.pyplot as pt #Import all the necessaries Libraries
import seaborn as sb
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
sb.set_style('whitegrid') #Sety the background Looks as White Grids
```

```python
x = np.linspace(0, 10, 30)
y = x ** 2
z = x ** 3
```

```python
fig = pt.figure()
axes = fig.add_axes([0, 0, 0.8, 0.8]) #An user defined backgrounds
axes.plot(x, y, c='r')
axes.set_title('y = x^2 Graph')
axes.set_xlabel('X-Axis')
axes.set_ylabel('Y-Axis')
pt.show()
```

```python
fig, axes = pt.subplots(nrows=2, ncols=1, figsize=(8, 2)) #Subplots
axes[0].plot(x, y)
axes[1].plot(y, x)
pt.tight_layout()
pt.show()
```

**Advanced Subplots**

```python
fig = pt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.3, 0.6, 0.34, 0.34])
axes1.set_title('Outer y=x**2 Graph')
axes2.set_title('Inner x=y**2 Graph')
axes1.plot(x, y, c='r', lw=2, linestyle='--')
axes2.plot(y, x, lw=4)
pt.show()
```

**SeaBorn Plotting**

```python
tips = sb.load_dataset('tips') #load the dataset to plot the statiscal graphs
tips.head()
```

```python
sb.displot(tips['total_bill'], kde=True) #Bar Plot with Kernal Density Estimation
pt.show()
```

```python
sb.jointplot(x='total_bill', y='tip', data=tips)
pt.show()
```

```python
sb.jointplot(x='total_bill', y='tip', data=tips, kind='hex') #Change the scatter points
pt.show()
```

```python
sb.pairplot(tips) #Multi Indexed Graph
pt.show()
```

**Categorical Plots**

```python
sb.barplot(x='sex', y='total_bill', data=tips)
pt.show()
```

```python
sb.countplot(x='sex', data=tips, hue='sex') #Counts the numbers of instances and plot them
pt.show()
```

```python
sb.boxplot(x='day', y='total_bill', data=tips, hue='sex')
pt.show()
```

```python
sb.violinplot(x='day', y='total_bill', data=tips, hue='smoker')
pt.show()
```

```python
sb.stripplot(x='day', y='tip', data=tips)
pt.show()
```

```python
sb.swarmplot(x='day', y='tip', data=tips, hue='sex')
pt.show()
```

```python
sb.catplot(x='day', y='total_bill', data=tips, kind='bar')
pt.show()
```

**Matrix plots**

```python
flights = sb.load_dataset('flights')
flights.head()
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```
