# Introduction to Neural Networks and Deep Learning

## What is Deep Learning?

Deep Learning is a branch of Artificial Intelligence (AI) and Machine Learning (ML) that teaches computers to learn from data using structures called Neural Networks. It is inspired by the human brain, where neurons work together to process information and make decisions.

Deep learning is used in:

* Image Recognition
* Voice Assistants
* Chatbots
* Self-driving Cars
* Recommendation Systems
* Medical Diagnosis

The main advantage of deep learning is that it can automatically learn patterns from large amounts of data without manual programming.

---

# What is a Neural Network?

A Neural Network is a collection of connected neurons organized into layers.

The three main layers are:

1. Input Layer
   Receives the data.

2. Hidden Layer
   Processes the information using mathematical calculations.

3. Output Layer
   Produces the final prediction or result.

Each connection between neurons has:

* Weights
* Biases
* Activation Functions

These values help the network learn patterns.

---

# How Neural Networks Learn

Neural networks learn by adjusting weights and biases during training.

The learning process includes:

1. Forward Propagation
2. Loss Calculation
3. Backpropagation
4. Weight Update

This process repeats many times until the model becomes accurate.

---

# Forward Propagation

In forward propagation:

* Input data enters the network.
* Calculations happen layer by layer.
* The model generates an output.

The neuron calculation is:

y=f(wx+b)

Where:

* (x) = input
* (w) = weight
* (b) = bias
* (f) = activation function
* (y) = output

---

# Loss Function

The Loss Function measures how wrong the prediction is.

Example:

* Actual Output = 1
* Predicted Output = 0.4

The model calculates the error and tries to reduce it.

Common loss functions:

* Mean Squared Error (MSE)
* Binary Cross Entropy
* Categorical Cross Entropy

---

# Backpropagation

Backpropagation is the most important learning process in neural networks.

It works by:

1. Calculating the error
2. Sending the error backward through the network
3. Updating weights to reduce mistakes

The model improves after every iteration.

Without backpropagation, the neural network cannot learn effectively.

---

# Model Training

Model training means teaching the neural network using data.

Training steps:

1. Load dataset
2. Build neural network
3. Compile model
4. Train using epochs
5. Evaluate performance

Important terms:

* Epoch: One complete pass through the dataset
* Batch Size: Number of samples processed together
* Learning Rate: Controls how fast the model learns

---

# TensorFlow

TensorFlow is an open-source deep learning library developed by Google.

It helps developers:

* Build neural networks
* Train AI models
* Perform predictions
* Work with deep learning easily

TensorFlow supports:

* CPU
* GPU
* Large-scale AI applications

---

# Sequential Model in TensorFlow

The Sequential model is the simplest way to create neural networks in TensorFlow.

It creates layers one after another in sequence.

Example structure:
Input → Hidden Layer → Output Layer

Sequential models are best for beginners and simple deep learning tasks.

---

# Dense Layer

A Dense layer is a fully connected neural network layer.

In a Dense layer:

* Every neuron connects to all neurons in the next layer.

Dense layers are commonly used in:

* Classification
* Prediction
* Pattern Recognition

Example:

* Dense(128)
* Dense(64)
* Dense(10)

The number represents neurons in that layer.

---

# Creating a Simple Neural Network Using TensorFlow

Example code:

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create model
model = Sequential([
    Dense(128, activation='relu', input_shape=(10,)),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train model
model.fit(X_train, y_train, epochs=10)
```

---

# Explanation of the Code

1. Sequential()
   Creates the neural network structure.

2. Dense()
   Adds fully connected layers.

3. activation='relu'
   Helps the network learn complex patterns.

4. sigmoid
   Used for binary classification outputs.

5. compile()
   Configures optimizer and loss function.

6. fit()
   Trains the model using data.

7. evaluate()
   Checks model performance.

# Activation Functions

Activation functions decide whether a neuron should activate.

Common activation functions:

* ReLU
* Sigmoid
* Softmax
* Tanh

ReLU formula:

f(x)=\max(0,x)

Sigmoid formula:

\sigma(x)=\frac{1}{1+e^{-x}}
