```python
import seaborn as sb
import numpy as np
import matplotlib.pyplot as pt
```

```python
tips=sb.load_dataset('tips') #loading the dataset to plot the regression plot
tips.head()
```

```python
sb.lmplot(x='total_bill', y='tip', data=tips) #Plotting the linear Regression Plot for best Fit Line to scatter points
pt.show()
```

```python
sb.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm') #Plotting the linear Regression Plot for best Fit Line to scatter points with different categories
pt.show()
```
